# Peso de la Evidencia (PdE)

El WoE es una medida de la fuerza de una variable independiente (o de una categor√≠a/intervalo de esa variable) para discriminar entre dos grupos en la variable dependiente (objetivo). En el contexto de riesgo, estos grupos son compuestos "Buenos" (clientes que no defaultaron/incumplieron) y "Malos" (clientes que s√≠ defaultaron/incumplieron).

```python
# Vamos a hacer un ejemplo de preprocesamiento para variables categ√≥ricas
# Para esto, comenzaremos estimando el peso de la evidencia (PDE)
# PdE = ln(%Buenos/%Malos)
# El peso de la evidencia indica el poder predictivo de una variable 
# independiente (home_ownership) en relaci√≥n con la variable dependiente (buenos_malos). 
# PdE se describe como una medida de la separaci√≥n de buenos y malos clientes.
# "Clientes malos" se refiere a los clientes que incumplieron con un pr√©stamo.
# y "Buenos clientes" se refiere a los clientes que pagaron el pr√©stamo.
# ¬øPara que hacemos esto? Para combinar categor√≠as
# ¬øPor qu√© combinar categor√≠as con PdE similares?
# Es porque las categor√≠as con PdE similar tienen casi la misma proporci√≥n 
# de buenos y malos. En otras palabras, el comportamiento de ambas 
# categor√≠as es el mismo.
```

El Peso de la Evidencia (Weight of Evidence - WoE) es una t√©cnica fundamental en el modelado de riesgo, especialmente en el sector financiero y el scoring de cr√©dito, aunque sus principios se aplican en cualquier problema de clasificaci√≥n binaria.

# Valor de la Informaci√≥n (VI)

El an√°lisis del valor de la informaci√≥n es una t√©cnica de exploraci√≥n de datos que ayuda a determinar qu√© variables de un conjunto de datos tienen poder *predictivo o influencia en el valor de una variable dependiente espec√≠fica*.

 Evaluando el peso de las categor√≠as dentro de la variable  sino de la variable en s√≠ VI = Sumatoria (PdE*(proporci√≥n de los buenos en la categor√≠a - proporci√≥n de los malos en la categor√≠a))

Es una m√©trica clave para la selecci√≥n de variables en el modelado de riesgo.

Aqu√≠ est√° la gu√≠a general de interpretaci√≥n del IV:

    IV < 0.02: La variable tiene poco o ning√∫n poder predictivo (in√∫til).

    0.02 <= IV < 0.1: La variable es un predictor d√©bil.

    0.1 <= IV < 0.3: La variable es un predictor medio (√∫til).

    0.3 <= IV < 0.5: La variable es un predictor fuerte.

    IV >= 0.5: La variable podr√≠a ser demasiado fuerte y podr√≠a indicar multicolinealidad, sobreajuste, o "fuga de datos" (data leakage) del objetivo al predictor. Se debe investigar a fondo.

# Automatizaci√≥n del Preprocesamiento de Variables Categ√≥ricas

```python
# No ser√≠a grandioso si pudieramos automatizar el proceso de las sesiones
# anteriores para calcular el PdE y VI para cualquier para de variables
# dependiente (categ√≥rica) e independiente
# Definamos la funci√≥n PdE_categorica que ejecutar√° todos los pasos que hicimos
# en las sesiones anteriores
```

```python
def PdE_categorica(df, var_categorica, df_var_dependiente):
    df = pd.concat([df[var_categorica], df_var_dependiente], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                 df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis=1)
    df = df.iloc[:, [0,1,3]]
    df.columns = [df.columns.values[0], 'observaciones', 'prop_bueno']
    df['proporcion_obs'] = df['observaciones']/df['observaciones'].sum()
    df['n_buenos'] = df['prop_bueno']*df['observaciones']
    df['n_malos'] = (1- df['prop_bueno'])*df['observaciones'] 
    df['prop_n_buenos'] = df['n_buenos']/df['n_buenos'].sum()
    df['prop_n_malos'] = df['n_malos']/df['n_malos'].sum() 
    df['PdE'] = np.log(df['prop_n_buenos']/df['prop_n_malos'])
    df = df.sort_values(['PdE'])
    df = df.reset_index(drop = True)
    df['delta_prop_n_buenos'] = df['prop_n_buenos'].diff().abs() 
    df['delta_PdE'] = df['PdE'].diff().abs() #Diferencia Absoluta entre categor√≠as
    df['VI'] = (df['prop_n_buenos'] - df['prop_n_malos']) * df['PdE']
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df['sumatoriaVI'] = df['VI'].sum()
    return df
    
    
```

# Preprocesando Variables Continuas

La mayor diferencia entre procesar estas variables continuas con las varibles categoricas es la siguiente:

Las variables continuas se ordenan por sus valores o rangos y no por su peso de la evidencia.

```python
# Copiamos y pegamos la funci√≥n PdE_discreta
# Comentamos las l√≠neas de codigo en las que orden√°mabamos los valores por PdE y reseteabamos los √≠ndices
```

```python
def PdE_continua(df, var_categorica, df_var_dependiente):
    df = pd.concat([df[var_categorica], df_var_dependiente], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                 df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis=1)
    df = df.iloc[:, [0,1,3]]
    df.columns = [df.columns.values[0], 'observaciones', 'prop_bueno']
    df['proporcion_obs'] = df['observaciones']/df['observaciones'].sum()
    df['n_buenos'] = df['prop_bueno']*df['observaciones']
    df['n_malos'] = (1- df['prop_bueno'])*df['observaciones'] 
    df['prop_n_buenos'] = df['n_buenos']/df['n_buenos'].sum()
    df['prop_n_malos'] = df['n_malos']/df['n_malos'].sum() 
    df['PdE'] = np.log(df['prop_n_buenos']/df['prop_n_malos'])
    # df = df.sort_values(['PdE'])
    # df = df.reset_index(drop = True)
    df['delta_prop_n_buenos'] = df['prop_n_buenos'].diff().abs() 
    df['delta_PdE'] = df['PdE'].diff().abs() #Diferencia Absoluta entre categor√≠as
    df['VI'] = (df['prop_n_buenos'] - df['prop_n_malos']) * df['PdE']
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df['sumatoriaVI'] = df['VI'].sum()
    return df
```

```python
# Como el resultado ser√° un df, no necesitamos hacer ning√∫n cambio en nuestra 
# funci√≥n de gr√°fico, la √∫nica diferencia ser√° que el gr√°fico no estar√° ordenado
# de menor a mayor PdE
```

# Preprocesando los datos de evaluaci√≥n

dividimos el conjunto de datos en dos partes, datos de entrenamiento y datos de evaluaci√≥n, training y testing en ingl√©s. 

Luego usamos el conjunto de datos de entrenamiento para generar variables dummies adecuadas para nuestro modelo estad√≠stico.  Pero para probar el modelo en el conjunto de datos de evaluaci√≥n, debemos tener las mismas 
variables dummies que las del conjunto de entrenamiento. ¬øC√≥mo podemos lograr esto?

Podemos llevar a cabo esto de forma muy sencilla reutilizando el c√≥digo que hemos escrito para la base training. 
Eso significa que podemos usar el mismo c√≥digo para crear las mismas variables dummies para el df de datos de evaluaci√≥n (testing)

Sin embargo, hay una cosa que debemos hacer antes de comenzar, ya que si solo corremos el c√≥digo
sobrescribir√≠amos las variables creadas para entrenar el modelo. Es por eso que antes de regresar 
al punto donde comenzamos nuestro preprocesamiento, es importante guardar el conjunto de datos 
training, tanto para las variables independientes como para la variable dependiente.

# Partiendo la base

```python
from sklearn.model_selection import train_test_split
```

### Cuando ocurre el Overfitting

Cuando esto sucede el modelo no puede funcionar con precesion contra datos no vistos, lo que anula su proposito inicial. La generalizacion de un modelo a nuevos datos es lo que permite usarlo para hacer predicciones. Cuando un modelo se entrena con dataset historicos puede que en el mismo este incluido datos de ruido "White-noise"

#### Como identificarlo

    1. Cuando el modelo se ajusta demasiado al conjunto de entrenamiento, el modelo se "Sobre ajusta" y no puede generalizar bien los nuevos datos.

    2. Bajas tasas de error y una alta vaariabza son buenos indicadores de sobre ajuste. 

```python

# 2. Define tus variables X (independientes) y y (dependiente)
# Es una buena pr√°ctica asignarlas a variables X e y para mayor claridad
X = df_dummies # Todas las columnas EXCEPTO 'buenos_malos'
y = df_objetivo             # La columna 'buenos_malos'

# 3. Realiza la divisi√≥n de los datos
# Asigna las cuatro salidas de train_test_split a las variables correspondientes
# Se recomienda usar test_size y random_state para reproducibilidad y control del tama√±o
independientes_entreno, independientes_evaluacion, dependiente_entreno, dependiente_evaluacion = train_test_split(
    X,
    y,
    test_size=0.2,   # Por ejemplo, 20% de los datos para evaluaci√≥n
    random_state=99  # Para asegurar que la divisi√≥n sea la misma cada vez
)


print("Dimensiones de X_train (independientes_entreno):", independientes_entreno.shape)
print("Dimensiones de X_test (independientes_evaluacion):", independientes_evaluacion.shape)
print("Dimensiones de y_train (dependiente_entreno):", dependiente_entreno.shape)
print("Dimensiones de y_test (dependiente_evaluacion):", dependiente_evaluacion.shape)
```

```python
# ¬°¬°El siguiente c√≥digo est√° comentado y s√≥lo har√° sentido una vez hayas preprocesado la base entrenamiento!!
# ¬°¬°Aqu√≠ vamos a sustituir las bases de entreno con las bases de evaluaci√≥n!!
# Esto con el objetivo de generar las mismas variables dummies en ambas bases utilizando
# el mismo c√≥digo que ya hab√≠amos escrito

#independientes_entreno = independientes_evaluacion
#dependiente_entreno = dependiente_evaluacion
```

```python
datos_guardados_independientes_entreno = independientes_entreno
```

```python
datos_guardados_dependiente_entreno = dependiente_entreno
```

```python
# Luego de volver a correr el c√≥digo con el df testing
# guardamos los df de evaluacion (testing)
```

```python
datos_guardados_independientes_evaluacion = independientes_evaluacion
```

```python
datos_guardados_dependiente_evaluacion = dependiente_evaluacion
```
