# Peso de la Evidencia (PdE)

El WoE es una medida de la fuerza de una variable independiente (o de una categoría/intervalo de esa variable) para discriminar entre dos grupos en la variable dependiente (objetivo). En el contexto de riesgo, estos grupos son compuestos "Buenos" (clientes que no defaultaron/incumplieron) y "Malos" (clientes que sí defaultaron/incumplieron).

```python
# Vamos a hacer un ejemplo de preprocesamiento para variables categóricas
# Para esto, comenzaremos estimando el peso de la evidencia (PDE)
# PdE = ln(%Buenos/%Malos)
# El peso de la evidencia indica el poder predictivo de una variable 
# independiente (home_ownership) en relación con la variable dependiente (buenos_malos). 
# PdE se describe como una medida de la separación de buenos y malos clientes.
# "Clientes malos" se refiere a los clientes que incumplieron con un préstamo.
# y "Buenos clientes" se refiere a los clientes que pagaron el préstamo.
# ¿Para que hacemos esto? Para combinar categorías
# ¿Por qué combinar categorías con PdE similares?
# Es porque las categorías con PdE similar tienen casi la misma proporción 
# de buenos y malos. En otras palabras, el comportamiento de ambas 
# categorías es el mismo.
```

El Peso de la Evidencia (Weight of Evidence - WoE) es una técnica fundamental en el modelado de riesgo, especialmente en el sector financiero y el scoring de crédito, aunque sus principios se aplican en cualquier problema de clasificación binaria.

# Valor de la Información (VI)

El análisis del valor de la información es una técnica de exploración de datos que ayuda a determinar qué variables de un conjunto de datos tienen poder *predictivo o influencia en el valor de una variable dependiente específica*.

 Evaluando el peso de las categorías dentro de la variable  sino de la variable en sí VI = Sumatoria (PdE*(proporción de los buenos en la categoría - proporción de los malos en la categoría))

Es una métrica clave para la selección de variables en el modelado de riesgo.

Aquí está la guía general de interpretación del IV:

    IV < 0.02: La variable tiene poco o ningún poder predictivo (inútil).

    0.02 <= IV < 0.1: La variable es un predictor débil.

    0.1 <= IV < 0.3: La variable es un predictor medio (útil).

    0.3 <= IV < 0.5: La variable es un predictor fuerte.

    IV >= 0.5: La variable podría ser demasiado fuerte y podría indicar multicolinealidad, sobreajuste, o "fuga de datos" (data leakage) del objetivo al predictor. Se debe investigar a fondo.

# Automatización del Preprocesamiento de Variables Categóricas

```python
# No sería grandioso si pudieramos automatizar el proceso de las sesiones
# anteriores para calcular el PdE y VI para cualquier para de variables
# dependiente (categórica) e independiente
# Definamos la función PdE_categorica que ejecutará todos los pasos que hicimos
# en las sesiones anteriores
```

```python
def PdE_categorica(df, var_categorica, df_var_dependiente):
    df = pd.concat([df[var_categorica], df_var_dependiente], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                 df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis=1)
    df = df.iloc[:, [0,1,3]]
    df.columns = [df.columns.values[0], 'observaciones', 'prop_bueno']
    df['proporcion_obs'] = df['observaciones']/df['observaciones'].sum()
    df['n_buenos'] = df['prop_bueno']*df['observaciones']
    df['n_malos'] = (1- df['prop_bueno'])*df['observaciones'] 
    df['prop_n_buenos'] = df['n_buenos']/df['n_buenos'].sum()
    df['prop_n_malos'] = df['n_malos']/df['n_malos'].sum() 
    df['PdE'] = np.log(df['prop_n_buenos']/df['prop_n_malos'])
    df = df.sort_values(['PdE'])
    df = df.reset_index(drop = True)
    df['delta_prop_n_buenos'] = df['prop_n_buenos'].diff().abs() 
    df['delta_PdE'] = df['PdE'].diff().abs() #Diferencia Absoluta entre categorías
    df['VI'] = (df['prop_n_buenos'] - df['prop_n_malos']) * df['PdE']
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df['sumatoriaVI'] = df['VI'].sum()
    return df
    
    
```

# Preprocesando Variables Continuas

La mayor diferencia entre procesar estas variables continuas con las varibles categoricas es la siguiente:

Las variables continuas se ordenan por sus valores o rangos y no por su peso de la evidencia.

```python
# Copiamos y pegamos la función PdE_discreta
# Comentamos las líneas de codigo en las que ordenámabamos los valores por PdE y reseteabamos los índices
```

```python
def PdE_continua(df, var_categorica, df_var_dependiente):
    df = pd.concat([df[var_categorica], df_var_dependiente], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                 df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis=1)
    df = df.iloc[:, [0,1,3]]
    df.columns = [df.columns.values[0], 'observaciones', 'prop_bueno']
    df['proporcion_obs'] = df['observaciones']/df['observaciones'].sum()
    df['n_buenos'] = df['prop_bueno']*df['observaciones']
    df['n_malos'] = (1- df['prop_bueno'])*df['observaciones'] 
    df['prop_n_buenos'] = df['n_buenos']/df['n_buenos'].sum()
    df['prop_n_malos'] = df['n_malos']/df['n_malos'].sum() 
    df['PdE'] = np.log(df['prop_n_buenos']/df['prop_n_malos'])
    # df = df.sort_values(['PdE'])
    # df = df.reset_index(drop = True)
    df['delta_prop_n_buenos'] = df['prop_n_buenos'].diff().abs() 
    df['delta_PdE'] = df['PdE'].diff().abs() #Diferencia Absoluta entre categorías
    df['VI'] = (df['prop_n_buenos'] - df['prop_n_malos']) * df['PdE']
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df['sumatoriaVI'] = df['VI'].sum()
    return df
```

```python
# Como el resultado será un df, no necesitamos hacer ningún cambio en nuestra 
# función de gráfico, la única diferencia será que el gráfico no estará ordenado
# de menor a mayor PdE
```

# Preprocesando los datos de evaluación

dividimos el conjunto de datos en dos partes, datos de entrenamiento y datos de evaluación, training y testing en inglés. 

Luego usamos el conjunto de datos de entrenamiento para generar variables dummies adecuadas para nuestro modelo estadístico.  Pero para probar el modelo en el conjunto de datos de evaluación, debemos tener las mismas 
variables dummies que las del conjunto de entrenamiento. ¿Cómo podemos lograr esto?

Podemos llevar a cabo esto de forma muy sencilla reutilizando el código que hemos escrito para la base training. 
Eso significa que podemos usar el mismo código para crear las mismas variables dummies para el df de datos de evaluación (testing)

Sin embargo, hay una cosa que debemos hacer antes de comenzar, ya que si solo corremos el código
sobrescribiríamos las variables creadas para entrenar el modelo. Es por eso que antes de regresar 
al punto donde comenzamos nuestro preprocesamiento, es importante guardar el conjunto de datos 
training, tanto para las variables independientes como para la variable dependiente.

# Partiendo la base

```python
from sklearn.model_selection import train_test_split
```

### Cuando ocurre el Overfitting

Cuando esto sucede el modelo no puede funcionar con precesion contra datos no vistos, lo que anula su proposito inicial. La generalizacion de un modelo a nuevos datos es lo que permite usarlo para hacer predicciones. Cuando un modelo se entrena con dataset historicos puede que en el mismo este incluido datos de ruido "White-noise"

#### Como identificarlo

    1. Cuando el modelo se ajusta demasiado al conjunto de entrenamiento, el modelo se "Sobre ajusta" y no puede generalizar bien los nuevos datos.

    2. Bajas tasas de error y una alta vaariabza son buenos indicadores de sobre ajuste. 

```python

# 2. Define tus variables X (independientes) y y (dependiente)
# Es una buena práctica asignarlas a variables X e y para mayor claridad
X = df_dummies # Todas las columnas EXCEPTO 'buenos_malos'
y = df_objetivo             # La columna 'buenos_malos'

# 3. Realiza la división de los datos
# Asigna las cuatro salidas de train_test_split a las variables correspondientes
# Se recomienda usar test_size y random_state para reproducibilidad y control del tamaño
independientes_entreno, independientes_evaluacion, dependiente_entreno, dependiente_evaluacion = train_test_split(
    X,
    y,
    test_size=0.2,   # Por ejemplo, 20% de los datos para evaluación
    random_state=99  # Para asegurar que la división sea la misma cada vez
)


print("Dimensiones de X_train (independientes_entreno):", independientes_entreno.shape)
print("Dimensiones de X_test (independientes_evaluacion):", independientes_evaluacion.shape)
print("Dimensiones de y_train (dependiente_entreno):", dependiente_entreno.shape)
print("Dimensiones de y_test (dependiente_evaluacion):", dependiente_evaluacion.shape)
```
